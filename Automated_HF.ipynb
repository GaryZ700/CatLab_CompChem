```python
from molecule import atom
from molecule import vector
from molecule import guassian
from molecule import molecule
from notebookImporter import importNotebook

HartreeFock = importNotebook("HartreeClass")

#set up the automated calculations
E = []
R = []
startR = 0.1
endR = 5
step = 0.01
basisName = "STO-3G"

delta = (endR - startR) / step
currentR = startR

while(not(currentR >= endR)):
    
    #define the molecular system
    system = molecule()
    system.addAtom(atom(vector(1,1,1+currentR), 1, 1))
    system.addAtom(atom(vector(1,1,1), 1, 1))
    system.addBasis(basisName)
    
    HF = HartreeFock.HF(system, pow(10, -15), 100)
    
    R.append(currentR)
    E.append(HF.SCF())
    
    #update the current intenuclear distance
    currentR += step
    
    print(currentR)
```

```python
from matplotlib import pyplot as plt

s = 0
e = 500
top, bottom = plt.ylim()
plt.ylim(-2, 0)
plt.plot(R[s:e], E[s:e], '-.')
plt.show()
print(R[(E.index(min(E)))])
```
