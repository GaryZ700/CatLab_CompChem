```python
from molecule import atom
from molecule import vector
from molecule import gaussian
from molecule import molecule
from notebookImporter import importNotebook

HartreeFock = importNotebook("HartreeClass")

#set up the automated calculations
E = []
R = []
MOEnergy = []
startR = 0.1
endR = 5
step = 0.01
basisName = "6-21G"
convergenceCritera = pow(10, -15)
maxIterations = 100

delta = (endR - startR) / step
currentR = startR

while(not(currentR >= endR)):
    
    #define the molecular system
    system = molecule()
    system.addAtom(atom(vector(5,5,2+currentR), 1, 1))
    system.addAtom(atom(vector(5,5,2), 1, 1))
    system.addBasis(basisName)
    
    #Set up the Hartree Procedure
    HF = HartreeFock.HF(system, convergenceCritera, maxIterations)

    #Store Bond distance, associated Energy, and MO Energies
    print("Bond Distance: " + str(currentR))
    R.append(currentR)
    E.append(HF.SCF())
    MOEnergy.append(HF.MOEnergy)
    print()
    
    #update the current intenuclear distance
    currentR += step
```

```python
#Prepare for energy level computation here

#import all packages 
import math
from matplotlib import pyplot as plt
from plotly.offline import iplot, init_notebook_mode

#connect plotly to the Juypter notebook
init_notebook_mode(connected=True)

#Define all functions to be used for this calculation here
#3D distance function using the pytagorean thereom
def distance(p1, p2):
    
    radicand = 0
    
    for dim, value in enumerate(p1):
        radicand += pow((p1[dim] - p2[dim]), 2)
        
    return math.sqrt(radicand)

#--------------------------------------------------------------------------------

#Returns values for the morse potential for all values of r specified in a list
def morsePotential(R, E, w):
    
    #declare list used to store the values for the approximation
    potential = []
    
    #compute all the requred parameters for the computation
    D = abs(min(E))
    r0 = R[E.index(min(E))]
    u = 0.5
    a = math.sqrt( 2 * u / D ) * w * math.pi
    
    for r in R:
         approximateE = (D*math.exp( (-2 * a) * (r - r0))) - ( 2*D*math.exp( -a * ( r - r0 )))
            
         #Other Form of the Morse Potential
         #approximateE = D * pow(1 - math.exp( -a * (r-r0) ), 2) - D
         
         potential.append(approximateE)
    
    return potential

#--------------------------------------------------------------------------------

#Function that estimates the correct W0, classical wave frequency, for the molecule
#using a form of the variational method
#Domain is the number of points back from the minmum energy of the well the method
#should go to begin computing the morse potential from
def computeMorsePotential(R, E, domain, w=0.02):

    #declare all variables here
    dw = 0.00001
    averageDistance = []
    bestApproximation = []
    add = True
    addSwitch = False
    
    #set the start index the specifed number of points behind the bottom of the well
    startIndex = E.index(min(E)) - domain

    #While loop to perform the variational portion of the computation
    while(True):
    
        distances = []
        
        #compute a morse potential 
        approximation = morsePotential(R, E, w)

        for index in range(2*domain + 1):
            currentIndex = startIndex + index
            #print("Current Index: " + str(currentIndex))
        
            distances.append( distance( [R[currentIndex], E[currentIndex]], [R[currentIndex], approximation[currentIndex]] ) )
            #print(distances[len(distances)-1])
            #print()
    
        averageDistance.append( sum(distances) / len(distances) )
        
        if(len(averageDistance)>=2):
            difference = averageDistance[len(averageDistance)-1] - averageDistance[len(averageDistance)-2]
            if(difference > 0):
                if(not addSwitch):
                    add = not add
                    addSwitch = True
                    bestApproximation = approximation
                else:
                    print("Computed Omega: " + str(w))
                    return approximation
        
        #print("Omega: " + str(w))
        #print()
        if(add):
            w += dw
        else:
            w -= dw
```

```python
#Compute a variationally optimized Morse Potential given the HF computed well
approximation = computeMorsePotential(R, E, 24)

#approximation = morsePotential(R, E)
minEIndex = E.index(min(E))
minE = E.pop(minEIndex)
optimalR = R.pop(minEIndex)

figure = { 
    "data": [
        
        #create the Hartree-Fock method generated graph
        {
            "type":"scatter",
            "x":R,
            "y":E,
            "connectgaps":False,
            "mode":"markers", 
            "name":"Hartree-Fock Computed",
            "marker":{"color":"blue"}
        },
    
        #highlight the minimum energy point in red
        {
            "type":"scatter",
            "x":[optimalR],
            "y":[minE],
            "name":"Optimal Bond Distance",
            "marker":{"color":"red"}
        },
        
        #create the morse potential approximation
        {
            "type":"scatter",
            "x":R, 
            "y":approximation,
            "connectgaps":True,
            "name":"Morse Potential Approximation",
            "marker":{"color":"green"}
        },
    ],
    "layout":
        {
           "xaxis":{"title":"Bond Distance in Atomic Units"},
           "yaxis":{"title":"Energy in Hartrees"},
            "title":{"text":"Hartree-Fock Energy VS Bond Distance"}
            },    
}

#add in energy levels to the figure
MOEnergy = MOEnergy[minEIndex]

#Loop over all MO Energies
for MOE in range(len(MOEnergy)):
        figure["data"].append(
            {
                "type":"scatter",
                "name":"MO Energy " + str(MOE+1),
                "x":[optimalR-0.4 + 0.4*i for i in range(3)],
                "y":[MOEnergy[MOE]]*3
            }
        )
    
iplot(figure)

#graph using matplotlib
#use inline to display graph in notebook
#use qt to display graph in a separate window
%matplotlib inline

plt.rcParams.update({'font.size': 25})

start = 0
end = len(R)
top, bottom = plt.ylim()
plt.plot(R[start:end], E[start:end], '-.')
#plt.show()

#print optimized bond distance
print("Optimal Bond Distance: " + str(R[(E.index(min(E)))]))
```
